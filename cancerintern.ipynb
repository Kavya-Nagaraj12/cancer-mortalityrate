```python
X = df.drop('TARGET_deathRate',axis=1)
y = df['TARGET_deathRate']
```

```python

from sklearn.model_selection import train_test_split #splitting training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y , train_size=0.75,test_size = 0.25, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler

#Import library for model building
from sklearn.linear_model import LinearRegression

scaler = StandardScaler() #feature scaling
X_train[X_train.columns] = scaler.fit_transform(X_train[X_train.columns])
X_train
```

```python
X_test[X_test.columns] =scaler.transform(X_test[X_test.columns])
```

```python
lr = LinearRegression() #fitting the linaer model
obj=lr.fit(X_train , y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred[0:5]
```

```python
y_test[0:5]
```

```python
obj.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor #randomforest model
```

```python
model_rfr = RandomForestRegressor(n_estimators=10000, n_jobs=-1,random_state=1)
obj1=model_rfr.fit(X_train , y_train)
y_pred = model_rfr.predict(X_test)
y_pred
```
